```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
yolov5xdata = pd.read_csv('runs/train/yolov5x_visdrone/results.csv', sep=',')
yolov5xdata.columns = yolov5xdata.columns.str.replace(' ', '')
```

```python
plt.plot(yolov5xdata['epoch'], yolov5xdata['train/cls_loss'], label="Training")
plt.plot(yolov5xdata['epoch'], yolov5xdata['val/cls_loss'], label="Validation")
plt.legend()
plt.xlabel("Epochs")
plt.ylabel("Loss")
plt.title("Yolo v5x Classification Loss")
plt.show()
```

```python
plt.plot(yolov5xdata['epoch'], yolov5xdata['train/box_loss'], label="Training")
plt.plot(yolov5xdata['epoch'], yolov5xdata['val/box_loss'], label="Validation")
plt.legend()
plt.xlabel("Epochs")
plt.ylabel("Loss")
plt.title("Yolo v5x Bounding Box Loss")
plt.show()
```

```python
plt.plot(yolov5xdata['epoch'], yolov5xdata['metrics/precision'], label="Precision")
plt.plot(yolov5xdata['epoch'], yolov5xdata['metrics/recall'], label="Recall")
plt.legend()
plt.xlabel("Epochs")
plt.ylabel("Precision/Recall")
plt.title("Yolo v5x Precision and Recall")
plt.show()
```

```python
ymax = max(yolov5xdata['metrics/mAP_0.5'])
xpos = list(yolov5xdata['metrics/mAP_0.5'].values).index(ymax)
xmax = yolov5xdata['epoch'][xpos]

plt.plot(yolov5xdata['epoch'], [ymax for i in range(len(yolov5xdata['epoch']))], '--',color='k')
plt.text(2, ymax-0.02, "mAP: "+str(ymax))

ymax = max(yolov5xdata['metrics/mAP_0.5:0.95'])
xpos = list(yolov5xdata['metrics/mAP_0.5:0.95'].values).index(ymax)
xmax = yolov5xdata['epoch'][xpos]

plt.plot(yolov5xdata['epoch'], [ymax for i in range(len(yolov5xdata['epoch']))], '--',color='k')
plt.text(2, ymax-0.02, "mAP: "+str(ymax))

plt.plot(yolov5xdata['epoch'], yolov5xdata['metrics/mAP_0.5'], label="mAP@0.5", color="tab:orange")
plt.plot(yolov5xdata['epoch'], yolov5xdata['metrics/mAP_0.5:0.95'], label="mAP@[0.5:0.95]",color='tab:orange')
plt.legend()
plt.xlabel("Epochs")
plt.ylabel("mAP")
plt.title("Yolo v5x mAP")
plt.show()
```
