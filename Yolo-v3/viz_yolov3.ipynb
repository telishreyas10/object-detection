```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
yolov3sdata = pd.read_csv('runs/detect/train3/results.csv', sep=',')
yolov3sdata.columns = yolov5sdata.columns.str.replace(' ', '')
```

```python
plt.plot(yolov3sdata['epoch'], yolov3sdata['train/cls_loss'], label="Training")
plt.plot(yolov3sdata['epoch'], yolov3sdata['val/cls_loss'], label="Validation")
plt.legend()
plt.xlabel("Epochs")
plt.ylabel("Loss")
plt.title("Yolo v3 Classification Loss")
plt.show()
```

```python
plt.plot(yolov3sdata['epoch'], yolov3sdata['train/box_loss'], label="Training")
plt.plot(yolov3sdata['epoch'], yolov3sdata['val/box_loss'], label="Validation")
plt.legend()
plt.xlabel("Epochs")
plt.ylabel("Loss")
plt.title("Yolo v3 Bounding Box Loss")
plt.show()
```

```python
plt.plot(yolov3sdata['epoch'], yolov3sdata['metrics/precision(B)'], label="Precision")
plt.plot(yolov3sdata['epoch'], yolov3sdata['metrics/recall(B)'], label="Recall")
plt.legend()
plt.xlabel("Epochs")
plt.ylabel("Precision/Recall")
plt.title("Yolo v3 Precision and Recall")
plt.show()
```

```python
ymax = max(yolov3sdata['metrics/mAP50(B)'])
xpos = list(yolov3sdata['metrics/mAP50(B)'].values).index(ymax)
xmax = yolov3sdata['epoch'][xpos]

plt.plot(yolov3sdata['epoch'], [ymax for i in range(len(yolov3sdata['epoch']))], '--',color='k')
plt.text(2, ymax-0.02, "mAP: "+str(ymax))

ymax = max(yolov3sdata['metrics/mAP50-95(B)'])
xpos = list(yolov3sdata['metrics/mAP50-95(B)'].values).index(ymax)
xmax = yolov3sdata['epoch'][xpos]

plt.plot(yolov3sdata['epoch'], [ymax for i in range(len(yolov3sdata['epoch']))], '--',color='k')
plt.text(2, ymax-0.02, "mAP: "+str(ymax))

plt.plot(yolov3sdata['epoch'], yolov3sdata['metrics/mAP50(B)'], label="mAP@0.5", color="tab:orange")
plt.plot(yolov3sdata['epoch'], yolov3sdata['metrics/mAP50-95(B)'], label="mAP@[0.5:0.95]", color="tab:orange")
plt.legend()
plt.xlabel("Epochs")
plt.ylabel("mAP")
plt.title("Yolo v3 mAP")
plt.show()
```
