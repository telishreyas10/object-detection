```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
yolov5sdata = pd.read_csv('runs/train/yolov5s_visdrone/results.csv', sep=',')
yolov5sdata.columns = yolov5sdata.columns.str.replace(' ', '')
```

```python
plt.plot(yolov5sdata['epoch'], yolov5sdata['train/cls_loss'], label="Training")
plt.plot(yolov5sdata['epoch'], yolov5sdata['val/cls_loss'], label="Validation")
plt.legend()
plt.xlabel("Epochs")
plt.ylabel("Loss")
plt.title("Yolo v5s Classification Loss")
plt.show()
```

```python
plt.plot(yolov5sdata['epoch'], yolov5sdata['train/box_loss'], label="Training")
plt.plot(yolov5sdata['epoch'], yolov5sdata['val/box_loss'], label="Validation")
plt.legend()
plt.xlabel("Epochs")
plt.ylabel("Loss")
plt.title("Yolo v5s Bounding Box Loss")
plt.show()
```

```python
plt.plot(yolov5sdata['epoch'], yolov5sdata['metrics/precision'], label="Precision")
plt.plot(yolov5sdata['epoch'], yolov5sdata['metrics/recall'], label="Recall")
plt.legend()
plt.xlabel("Epochs")
plt.ylabel("Precision/Recall")
plt.title("Yolo v5s Precision and Recall")
plt.show()
```

```python
ymax = max(yolov5sdata['metrics/mAP_0.5'])
xpos = list(yolov5sdata['metrics/mAP_0.5'].values).index(ymax)
xmax = yolov5sdata['epoch'][xpos]

plt.plot(yolov5sdata['epoch'], [ymax for i in range(len(yolov5sdata['epoch']))], '--',color='k')
plt.text(2, ymax-0.02, "mAP: "+str(ymax))

ymax = max(yolov5sdata['metrics/mAP_0.5:0.95'])
xpos = list(yolov5sdata['metrics/mAP_0.5:0.95'].values).index(ymax)
xmax = yolov5sdata['epoch'][xpos]

plt.plot(yolov5sdata['epoch'], [ymax for i in range(len(yolov5sdata['epoch']))], '--',color='k')
plt.text(2, ymax-0.02, "mAP: "+str(ymax))

plt.plot(yolov5sdata['epoch'], yolov5sdata['metrics/mAP_0.5'], label="mAP@0.5", color="tab:orange")
plt.plot(yolov5sdata['epoch'], yolov5sdata['metrics/mAP_0.5:0.95'], label="mAP@[0.5:0.95]", color="tab:orange")
plt.legend()
plt.xlabel("Epochs")
plt.ylabel("mAP")
plt.title("Yolo v5s mAP")
plt.show()
```
